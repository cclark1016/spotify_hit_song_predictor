```python
#ID List function

def get_playlist_tracks(username,playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    global track_id_list
    track_id_list = []
    global track_id_list_unfiltered
    track_id_list_unfiltered =[]
    for i in range(len(tracks)):
        track_id_list_unfiltered.append(tracks[i]['track']['id'])
    #Filter nonetypes out
    for val in track_id_list_unfiltered:
        if val != None:
            track_id_list.append(val)

#Track features function

def getTrackFeatures(id):
  meta = sp.track(id)
  features = sp.audio_features(id)

  # meta
  name = meta['name']
  album = meta['album']['name']
  artist = meta['album']['artists'][0]['name']
  release_date = meta['album']['release_date']
  length = meta['duration_ms']
  popularity = meta['popularity']
  explicit = meta['explicit']


  # features
  acousticness = features[0]['acousticness']
  danceability = features[0]['danceability']
  energy = features[0]['energy']
  instrumentalness = features[0]['instrumentalness']
  liveness = features[0]['liveness']
  loudness = features[0]['loudness']
  speechiness = features[0]['speechiness']
  tempo = features[0]['tempo']
  time_signature = features[0]['time_signature']
  valence = features[0]['valence']
  key = features[0]['key']
  mode = features[0]['mode']
  song_uri = id
  

  track = [name, album, artist, song_uri, release_date, length, popularity, danceability, acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, time_signature, explicit, valence, key, mode]
  return track
```
